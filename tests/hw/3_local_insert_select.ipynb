## 카테고리 별 검색 함수

```python
def select_want(watch, find) :
    sql_select = f"SELECT {watch} FROM try2_article WHERE category LIKE '%{find}%';"
    cursor.execute(sql_select)
    rows = cursor.fetchall()
    for row in rows:
        print(row)
```

```python
select_want('article_url','부동산')
```

```python
def select_want(category_name, *args):
    watch = ', '.join(args)
    sql_select = f"SELECT {watch} FROM try2_article WHERE category LIKE '%{category_name}%';"
    cursor.execute(sql_select)
    rows = cursor.fetchall()
    for row in rows:
        print(row)
```

```python
select_want('생활경제', 'article_url', 'title', 'author')
```
